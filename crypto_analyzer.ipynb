# Classification

## Predicting positive and negative outcome fails under binary classification problems. Classifier implemented classifier module is generic outcome classifier which, at the moment, perform one step ahead prediction. For our collected data, sample time is 1 minute(to better 'catch' system dynamics), so classifier predicts outcome sign one 1 ahead. This time step is not explicitly integrated in classifier and depends on data provided. On the other hand, 1-step prediction is classifier decision, but, these can be easily changed in order to address n-steps ahead prediction.  The most important step in classifier are features, which are generated using initial 'close' and 'volume' data. The 'close' feature is transformed in 'returns' to create target data(labels) for prediction, but also, to create base for creation predictive variables (features). Normalized volume is the only feature which is not generated from returns, it is correlated with other features but can provide additional insight into momentum. Returns values are used for further feature expansion, through returns of past records, and exponentially weighted moving average of returns.  Feature generation provides these values with different window sizes, to increase models possibily of recognizing present pattern in aggregated data. However, for actual model performance, it is important to select only variables caring significant pattern information. Because of that, feature selection is performed. Due to correlation among almost all  features, methodology used is creating predictive model as feature selector - model usage is to provide information which combination of variables leadsto higher model accuracy. Model used as feature selector is random forest which have strong ability of capturing feature importance through tree purity. After that, actual classifier is trained using only selected features from previous classifier. So, our classifier pipeline  (implemented inside CryptoClassifier object) consists of: - feature generation, - feature selection, - classifier training.  After training, all trained models are saved, so that classifier can be used in real time, on new data.   Basic usage can be observed through example for ETH coin.

## All collected data (1 min ticks for 10 days) is used in training process. However, using modified cross validation for time-series objects, we can gain thrustworthy model evaluation. Basic idea is to split train data on windowed folds from different , expanding, size, each fold split on train and test data and observe system prediction accuracy . This way we can see how is our system performing on different, not previously seen data. From the last accuracy, we see that our model was capable of learning all paterns provided in data, gaining accuracy of 100%. However, real system performance on unseen data is average of accuracy gained on each fold, which is arround 66%. This value is pretty low from the model validation point of view. However, model is demosntrating ability of catching patterns and there are lots of things that can be improved to increase system's accuracy. The most important one is training on bigger dataset. Another thing is creating additional (or dropping some) features using more domain knowledge. Last step is different final classifier. At the moment, RandomForest(with smaller number of estimators to avoid overfitting) is used. Better classifier for this usage would be SVM classifier, however it was not suitable for collected dat due to imbalanced dataset. Also, CryptoClassifier is at the moment developed as adaptable classifier(parameters are not tuned on specific data), so, trainig SVM classifier with automated hyperparameter would take lots of time and is not suitable for this demo purposes.
